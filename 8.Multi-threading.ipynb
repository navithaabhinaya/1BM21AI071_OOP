```python
"""Write a python program that implements a multi-thread application that has
three threads. First thread generates random integer every 1 second and if the
value is even, second thread computes the square of the number and prints. If
the value is odd, the third thread will print the value of cube of the number."""
import threading
import random
import time
import queue

def random_number_generator(queue):
    while True:
        number = random.randint(1, 100)
        print(f"Generated number: {number}")
        queue.put(number)
        time.sleep(1)

def square_calculator(queue):
    while True:
        number = queue.get()
        if number % 2 == 0:
            square = number ** 2
            print(f"Square of {number}: {square}")
        time.sleep(1)

def cube_printer(queue):
    while True:
        number = queue.get()

        if number % 2 != 0:
            cube = number ** 3
            print(f"Cube of {number}: {cube}")
        time.sleep(1)

# Create a shared queue
shared_queue = queue.Queue()

# Create threads
generator_thread = threading.Thread(target=random_number_generator, args=(shared_queue,))
square_thread = threading.Thread(target=square_calculator, args=(shared_queue,))
cube_thread = threading.Thread(target=cube_printer, args=(shared_queue,))

# Start threads
generator_thread.start()
square_thread.start()
cube_thread.start()

# Wait for threads to finish
generator_thread.join()
square_thread.join()
cube_thread.join()
```
